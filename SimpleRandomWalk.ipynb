# Simple Random Walk

The aim of this notebook is:
1. Generate a simple random walk.
2. Visualize the resulting path.
3. Understand how randomness accumulates over time.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(46)
```

## Step 1: Generate Random Steps

We start by generating a sequence of random steps. Each step will be either **+1** (up) or **-1** (down).

```python
# Number of steps
n_steps = 10000

# Generate random steps: +1 or -1 with equal probability
steps = np.random.choice([-1, 1], size=n_steps)

# Print first 10 steps
print("The array of 10 first steps:", steps[:10])
```

## Step 2: Compute the Cumulative Sum

The position at any time $t$ is simply the total of all the random steps taken up to that point.

Let $X_t$  denote the **position** of the walker at time $t$,
and $\varepsilon_t$ denote the **step** taken at time $t$, where:

$$
\varepsilon_t =
\begin{cases}
+1 & \text{with probability } 0.5 \\
-1 & \text{with probability } 0.5
\end{cases}
$$

Then, the position at time $t$ is given by the **cumulative sum** of all previous steps:

$$
X_t = \sum_{i=1}^{t} \varepsilon_i
$$

subject to:

$$
X_0 = 0
$$


```python
# Compute cumulative position at each step
position = np.cumsum(steps)

# Display first few positions
print("The array of 10 first positions is:", position[:10])
```

## Step 3: Visualize the Random Walk using Matplotlib.

- x-axis: number of steps.
- y-axis: the walkerâ€™s position.


```python
# Plot the random walk
plt.figure(figsize=(10, 6))
plt.plot(position, color='royalblue', linewidth=1.5)
plt.title( "Simple Random Walk", fontsize=16)
plt.xlabel("Step number", fontsize=12)
plt.ylabel("Position", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
